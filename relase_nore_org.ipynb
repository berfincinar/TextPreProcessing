```python
import texthero as hero
import pandas as pd
from texthero import preprocessing
```

```python
cols = [6]
df1= pd.read_excel('Release_Notes.xlsx',usecols=cols,sheet_name=['Sayfa1'])
```

```python
dfc=df1.copy()
dfc=pd.concat(df1[frame] for frame in df1.keys())
dfc
```

```python
df=pd.DataFrame()
df['Note']=dfc['Note']
```

```python
%%time
sq = []
for x in range(100):
    sq.append(x**2)

custom_pipeline = [preprocessing.fillna,
                   preprocessing.lowercase,
                   preprocessing.remove_whitespace,
                   preprocessing.remove_stopwords,
                   preprocessing.remove_diacritics,
                   preprocessing.remove_punctuation,
                   preprocessing.remove_digits]

df['Note'] = hero.clean(df['Note'], custom_pipeline)
df
```

```python
%%time
sq = []
for x in range(100):
    sq.append(x**2)


from nltk.stem import LancasterStemmer
lancaster=LancasterStemmer()

df['lancaster_clean']=df['Note'].apply(lambda x :" ".join(lancaster.stem(x) for x in x.split()))


```

```python
%%time
sq = []
for x in range(100):
    sq.append(x**2)

from porter2stemmer import Porter2Stemmer
stemmer = Porter2Stemmer()

df['porter2_clean']=df['Note'].apply(lambda x :" ".join(stemmer.stem(x) for x in x.split()))
```

```python
%%time
sq = []
for x in range(100):
    sq.append(x**2)


from nltk.stem import PorterStemmer 
st=PorterStemmer()

df['porter_clean']=df['Note'].apply(lambda x :" ".join(st.stem(x) for x in x.split()))
```

```python
%%time
sq = []
for x in range(100):
    sq.append(x**2)
    
import nltk
from nltk.stem import SnowballStemmer
stemmer_snb = SnowballStemmer("english")

df['snowball_clean']=df['Note'].apply(lambda x :" ".join(stemmer_snb.stem(x) for x in x.split()))
```
